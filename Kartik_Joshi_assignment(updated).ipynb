### Task 2. Nodes labeling (0 points)

Nodes in many real-world graphs represents some objects, say people in social networks or atoms in molecular networks. Usually, names of such nodes is in the string format, but in many cases it can be useful to make names and indices of nodes equal. Using such labeling, we can refer to a particular node by its index.

Write a function `convert_node_names_to_int` that takes a graph with string names and returns a new graph with integer names so that first node in the initial graph is called 0 in the new graph, second is 1 and so on. Also, save initial names in the node attribute `name`.

````python
def convert_node_names_to_int(G):
    # YOUR CODE HERE
    name_mapping = {}
    for idx, node in enumerate(G.nodes):
      name_mapping[node] = idx

    inv_map = {j:i for i, j in name_mapping.items()}

    newG = nx.Graph()

    newG.add_nodes_from(range(len(G)))

    for i, j in G.edges:
      newG.add_edge(name_mapping[i], name_mapping[j])

    for i in range(len(G)):
      newG.nodes[i]['name'] = inv_map[i]
    return newG


    #raise NotImplementedError()
````

Write a function `convert_to_digraph` that takes an undirected graph and returns a new directed graph where the direction is defined by the rule: the node $i$ follows to $j$ if the length of `name` of $i$ is at least the length of `name` of $j$. Keep the ordering of nodes and the node attribute `name`.

````python
def convert_to_digraph(G):
    # YOUR CODE HERE
    newG = nx.DiGraph()
    newG.add_nodes_from(G.nodes)
    for node in newG.nodes:
      newG.nodes[node]['name'] = G.nodes[node]['name']

    for i, j in G.edges:
      if len(G.nodes[i]['name']) > len(G.nodes[j]['name']):
        newG.add_edge(i, j)
      elif len(G.nodes[i]['name']) == len(G.nodes[j]['name']):
        newG.add_edge(i, j)
        newG.add_edge(j, i)
      else:
        newG.add_edge(j, i)
    return newG
    #raise NotImplementedError()
````

Your task is to draw the matrix graph using the kamada kawai layout with nonlinear modification to obtain a denser layout away from the NEO node. Let us define a proximity coefficient using Gaussian kernel

$$\text{coef}(i) = \exp\left(-\frac{\text{distance}^2}{2\sigma^2}\right)$$

where a _distance_ is the Euclidean distance from the node $i$ to the NEO node, $\sigma$ is the constant. Then the nodes should be moved towards the NEO node such that

$$\text{distance} = \text{coef}(i) \times \text{distance}$$

The simplest way to do it is to center the layout so that the NEO node is in x=0 y=0 and then to multiply all node positions by the proximity coefficient.

Write a function `dense_neo_layout` that takes a layout, sigma and returns a new (dense) layout.

_Hint: use `np.linalg.norm` to calculate Euclidean norm._

````python
def dense_neo_layout(layout, sigma):
    # YOUR CODE HERE
    import numpy as np

def dense_neo_layout(layout, sigma):
    # Extract the NEO node position
    neo_pos = layout['NEO']

    # Initialize a dictionary to store the new positions
    new_layout = {}

    # Calculate and apply modifications to node positions
    for node, pos in layout.items():
        # Calculate Euclidean distance from node to NEO
        distance = np.linalg.norm(np.array(pos) - np.array(neo_pos))

        # Calculate proximity coefficient using Gaussian kernel
        coef_i = np.exp(-distance ** 2 / (2 * sigma ** 2))

        # Modify node position based on proximity coefficient
        new_pos = np.array([coef_i * coord for coord in pos])


        # Store the new position in the dictionary
        new_layout[node] = new_pos

    # Center the layout so that NEO is at (0, 0)
    new_layout['NEO'] = np.array([0, 0])

    # Debugging statement
    # print("Positions of all nodes in the new layout:")
    # for node, pos in new_layout.items():
    #     print(node, ":", pos)

    return new_layout

    raise NotImplementedError()
````

Write a function `drawing_nodes_params` that takes a graph and returns a dictionary with parameters for `nx.draw_networkx` of the form `{'G': graph, 'node_color': color, ...}`. Parameters should contain a graph, kamada kawai layout, node colors from `viz.color` attribute. The colors should be converted from RGB range 0-255 to RGB range 0-1. The node size is a node degree `G.degree[node]` multiplied by 100.

````python
import numpy as np
import networkx as nx

def drawing_nodes_params(graph):
    # Initialize parameters dictionary
    params = {'G': graph}

    # Compute Kamada-Kawai layout
    layout = nx.kamada_kawai_layout(graph)

    # Extract node colors and sizes
    node_colors = []
    node_sizes = []
    for node, data in graph.nodes(data=True):
        # Extract color from node attribute
        color = (data['viz']['color']['r'] / 255,
                 data['viz']['color']['g'] / 255,
                 data['viz']['color']['b'] / 255)
        node_colors.append(color)

        # Calculate size based on node degree
        size = graph.degree[node] * 100
        node_sizes.append(size)

    # Convert node_colors to numpy array for rounding
    node_colors = np.array(node_colors)
    node_colors = np.round(node_colors, 2)

    # Add node colors and sizes to params dictionary
    params['node_color'] = node_colors
    params['node_size'] = node_sizes

    # Add Kamada-Kawai layout to params dictionary
    params['pos'] = layout

    return params
````

Your task is to draw the matrix graph where an edge width is 0 if a weight is 1, otherwise it is a weight value itself. An edge color is `red` if weight is at least 5 and `black` otherwise.

````python
def drawing_edges_params(graph):
    # Initialize parameters dictionary
    params = {'G': graph}

    # Compute Kamada-Kawai layout
    layout = nx.kamada_kawai_layout(graph)

    # Extract edge widths and colors
    edge_widths = []
    edge_colors = []
    for u, v, data in graph.edges(data=True):
        # Extract width based on edge weight
        width = 0 if data['weight'] == 1 else data['weight']
        edge_widths.append(width)

        # Determine edge color based on weight
        color = 'red' if data['weight'] >= 5 else 'black'
        edge_colors.append(color)

    # Add edge widths and colors to params dictionary
    params['width'] = edge_widths
    params['edge_color'] = edge_colors

    # Add Kamada-Kawai layout to params dictionary
    params['pos'] = layout

    return params
````

Write a function `labels_list_parameters` that takes a graph, layout and returns a list of parameters for repeated drawing labels using `nx.draw_networkx_labels`. Each element of the list is a dictionary that can be passed to `nx.draw_networkx_labels`. The first dictionary is for the node degree 6, the second is for 7 and so on.

````python
import math

def labels_list_parameters(graph, layout):
    # Step 1: Initialize an empty set to store visible degrees.
    visible_degrees = set()

    # Step 2: Iterate through each node in the graph.
    for node in graph.nodes():
        # Step 2a: If the degree of the node is less than 6, skip to the next node.
        if graph.degree(node) < 6:
            continue
        # Step 2b: Otherwise, add the degree to the set of visible degrees.
        visible_degrees.add(graph.degree(node))

    # Step 3: Convert the set of visible degrees to a list and sort it in ascending order.
    visible_degrees = sorted(list(visible_degrees))
    output_list = []  # Step 4: Initialize an empty list to store dictionaries containing parameters for each visible degree.

    # Step 5: Iterate through each visible degree.
    for degree in visible_degrees:
        # Step 5a: Create a dictionary containing parameters for the current degree.
        params = {
            'G': graph,
            'pos': layout,
            'labels': {},  # An empty dictionary to store node labels.
            'font_size': degree
        }
        # Step 5b: Append the dictionary to the output list.
        output_list.append(params)

    # Step 6: Create a dictionary that maps each visible degree to its index in the output list.
    degree_index_map = {degree: idx for idx, degree in enumerate(visible_degrees)}

    # Step 7: Iterate through each node in the graph.
    for node in graph.nodes():
        degree = graph.degree(node)
        # Step 7a: If the degree of the node is less than 6, skip to the next node.
        if degree < 6:
            continue
        # Step 7b: Determine the index of the current degree in the output list.
        index = degree_index_map[degree]
        # Step 7c: Add the node label to the corresponding dictionary in the output list.
        output_list[index]['labels'][node] = node
        # Step 7d: Adjust the y-coordinate of the node's position by subtracting 0.025 times the square root of the degree.
        layout[node][1] -= 0.025 * math.sqrt(degree)

    # Step 8: Return the list containing dictionaries of parameters for drawing labels at each visible degree.
    return output_list
````

Let us adjust a visible area and combine previous drawing techniques to get a more meaningful figure:
* take a node color from the `viz.color` node attribute converting 0-255 range to 0-1 range
* set a node size as $5\times\text{degree}^2$
* take an edge width from the `weight` edge attribute
* set an edge alpha is 0.2
* labels are invisible if a node degree less than 6
* set a font size as $3 \times \text{degree}^{2/3}$
* set a visible area such that a lower border is NEO, higher is AGENT SMITH, left is MORPHEUS, right is TANK.

Write a function `get_all_params` that takes a graph and layout and returns a dictionary with keys:
* `nodes_params` — dict, parameters for `nx.draw_networkx_nodes`
* `edges_params` — dict, parameters for `nx.draw_networkx_edges`
* `labels_param_list` — list of dict, parameters for `nx.draw_networkx_edges`, for degrees 6, 7 and so on
* `area` — list of float vlues: a left, right, lower, higher borders of the visible area

````python
def get_all_params(graph, layout):
    # YOUR CODE HERE
    out = {}

    # node parameters
    out['nodes_params'] = {
        'G': graph,
        'pos': layout,
        'node_color': [],
        'node_size': []
    }
    for n, data in graph.nodes(data=True):
        out['nodes_params']['node_color'].append(
            np.array(
                [data['viz']['color']['r']/255,
                 data['viz']['color']['g']/255,
                 data['viz']['color']['b']/255]
            )
        )
        out['nodes_params']['node_size'].append(
            5 * data['Degree']**2
        )

    # edge parameters
    out['edges_params'] = {
        'G': graph,
        'pos': layout,
        'alpha': 0.2,
        'width': [],
    }
    for data in graph.edges(data=True):
        out['edges_params']['width'].append(
                data[2]['weight']
        )


    # label parameters
    degrees = set()
    for n, deg in graph.degree:
        if deg < 6:
            continue
        degrees.add(deg)
    degrees = sorted(list(degrees))

    out['labels_param_list'] = []
    for d in degrees:
        out['labels_param_list'].append({
            'G': graph,
            'pos': layout,
            'labels': {},
            'font_size': 3 * d**(2/3)
        })
    deg_pos = dict(zip(degrees, list(range(len(degrees))) ))

    for n, data in graph.nodes(data=True):
        if data['Degree'] < 6:
            continue
        out['labels_param_list'][deg_pos[d]]['labels'][n] = n


    out['area'] = [
        layout['MORPHEUS'][0],
        layout['TANK'][0],
        layout['NEO'][1],
        layout['AGENT SMITH'][1]
    ]

    return out

    raise NotImplementedError()
````
